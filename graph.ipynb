```python
# first install pkg- pip install networkx & matplotlib
import networkx as nx
import matplotlib.pyplot as plt

g= nx.Graph()
g.add_nodes_from(["A", "B", "C", "D","E"])
edges=[("A", "B"), ("B", "C"), ("C", "D"), ("B", "D"), ("C","E"),("E", "A")]

g.add_edges_from(edges)

plt.figure(figsize=(8,6))
nx.draw(g, with_labels=True, node_color="red", node_size=1000, edge_color="gray", font_size=15 ,font_color="white", font_weight="bold")

plt.title("git add README.md using python")
plt.show()
```
